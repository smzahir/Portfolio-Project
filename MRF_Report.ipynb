```python
#importing necessary libraries

import numpy as np
import pandas as pd
```

```python
#reading the excel file

data = pd.read_excel('Final_Sales.xlsx')
```

```python
#converting the data into dataframe which makes the data more readable like excel
df = pd.DataFrame(data)
df.head()
```

```python
# finding which all columns conatins empty values so that columns with most empty will be eliminated

df.isnull().sum()
```

```python
# slicing or selecting and reordering columns based on its value
x = df.iloc[:,[1,3,5,9,19,22,23,27,28,33]]
x.head()
```

```python
# The total purchase rate is being calculated by summing quantity with unit price of item

x['Total'] = x['Billed Quantity'] * x['Purchase Rate']
```

```python
x
```

```python
# renaming the columns

y = x.rename(columns = {'Voucher Number':'Bill Number',
                        'Party Name':'Customer Name',
                        'Margin':'Profit',
                        'Total':'Total Purchase Rate',
                        'Billed Quantity':'Quantity',
                        'Item Name':'Item ID'})
```

```python
y
```

```python
# slicing or reordering based on priority

final_data = y.iloc[:,[0,1,2,3,9,4,6,10,7,8]]
final_data.head()
```

```python
# finding out the unique items in item group column
# the item group items will be converted into four groups
final_data['Item Group'].unique()
```

```python
# the items which fall under Car tyre is being seperated or named as car
final_data['Item Group'] = final_data['Item Group'].replace(['RADIALS','SCV 4W RADIAL 12/13','JEEP RADIALS','SMALL CAR 145 - 13',
                                               'MID CAR 14 - HIGH','MID CAR 14 - REGULAR','MINI CAR/VAN RADIALS',
                                               'SMALL CAR 155 - 13','PASS RADIAL PREMIUM','MID CAR 14 - POPULAR',
                                               'PASS RADIAL LUXURY','PASS RADIAL - REGULAR - 15','RCV RADIAL STD'],'CAR')

                                                
```

```python
# the item which fall under commercial vehicles is seperated or named as commercial
final_data['Item Group'] = final_data['Item Group'].replace(['LCV RIB','PICK UP BIAS','HCV BIAS','UV BIAS','SMALL PICK UP BIAS',
                                                             'SCV 4W BIAS 12/13','ICV BIAS','SCV 4W RADIAL 12/13','PICK UP RADIALS',
                                                             '7 50 16 LUG','HCV RADIAL','TRACTOR FRONT', '8 25 16 LUG','TRACTOR TRAILER']
                                                              ,'COMMERCIAL')
```

```python
# the item which fall under Three wheeled vehicles is seperated or named as Auto
final_data['Item Group'] = final_data['Item Group'].replace('SCV 3W BIAS','AUTO')
```

```python
# the item which fall under Scooter is seperated or named as Scooter
final_data['Item Group'] = final_data['Item Group'].replace(['SCOOTER - TT','SCOOTER - TL'],'SCOOTER')
```

```python
# the item which fall under bikes is seperated or named as bikes
final_data['Item Group'] = final_data['Item Group'].replace(['MCB - TT','MC - R','MCB - TL','MC - S','MC NV'],'BIKE')
```

```python
# the final data consists of flap and tubes which doesn't cause any impact to our data
final_data
```

```python
# list of four groups

Group = ['CAR','BIKE','SCOOTER','AUTO','COMMERCIAL']
```

```python
#finally the data is being filtered based on four groups which makes data crisp
new_data = final_data[final_data['Item Group'].isin(Group)]
```
